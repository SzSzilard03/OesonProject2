```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
df = pd.read_excel('Project2.xlsx')
data_array = df.to_numpy()

df.info()
```

```python
sns.countplot(data=df, x="work_year")
plt.title('Amount of jobs per year')
plt.xlabel('Year')
plt.ylabel('Count')
```

```python
unique_job_categories = df['job_category'].unique()

for category in unique_job_categories:
    category_df = df[df['job_category'] == category]
    mean_salary_by_job_title = category_df.groupby('job_title')['salary'].mean()
    plt.figure(figsize=(10, 6))
    mean_salary_by_job_title.plot(kind='bar')
    plt.title(f'Mean Salary by Job Title in {category}')
    plt.ylabel('Mean Salary')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
sns.barplot(data=df, x='work_year', y='salary', estimator=np.mean)
plt.title('Average Salary by Work Year')
plt.xlabel('Work Year')
plt.ylabel('Average Salary')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='salary_currency', y='salary_in_usd', estimator=np.median)
plt.title('Median Salary in USD by Salary Currency')
plt.xlabel('Salary Currency')
plt.ylabel('Median Salary in USD')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(23, 6))
sns.barplot(data=df, x='employee_residence', y='salary_in_usd', estimator=np.mean)
plt.title('Average Salary in USD by Employee Residence')
plt.xlabel('Employee Residence')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='experience_level', y='salary_in_usd', estimator=np.mean)
plt.title('Average Salary in USD by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='work_setting', y='salary_in_usd', hue='experience_level', estimator=np.mean)
plt.title('Average Salary in USD by Work Setting and Experience Level')
plt.xlabel('Work Setting')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Experience Level', bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='job_category', y='salary_in_usd', estimator=np.mean)
plt.title('Average Salary in USD by Job Category')
plt.xlabel('Job Category')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
work_setting_counts = df['work_setting'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(work_setting_counts, labels=work_setting_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Work Settings')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
